```python
%matplotlib widget
from webbrowser import get
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.animation as animation
from IPython import display
import time
from matplotlib.widgets import Slider, TextBox
import os
```

```python

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.35)
x = np.linspace(-10, 10, 10000)
y = np.sin(2*np.pi*x*1)
l, = plt.plot(x, y)
l.set_label('sin')
# ax.legend()

ax_ts = plt.axes([0.25, 0.15, 0.40, 0.03])
# ax_box = plt.axes([0.25, 0.10, 0.40, 0.03])
# ax_message = plt.axes([0.25, 0.05, 0.60, 0.03])

ts_num = Slider(ax_ts, 'freq', 1, 10, 1)

def get_num_slider(val):
    l.set_ydata(np.sin(2*np.pi*x*val))

ts_num.on_changed(get_num_slider)
plt.show()
```
