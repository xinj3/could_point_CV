```python
import numpy as np
from matplotlib import pyplot as plt
import json
```

```python
with open("radar_data-1666552189.4294128.log", "r") as f:
    file = f.read().replace("'", '"').split("\n")[:-1]
    data_dict = [json.loads(t) for t in file]
```

```python
data = []

for t in data_dict:
    cur = []
    for key in t["dataFrame"]["detected_points"].keys():
        cur.append([t["dataFrame"]["detected_points"][key]["x"], t["dataFrame"]["detected_points"][key]["y"], t["dataFrame"]["detected_points"][key]["z"]])
    data.append(np.array(cur))
```

```python
data = np.array(data)

frame_id = np.random.randint(0, np.shape(data)[0])
print(frame_id)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(data[frame_id][:, 1], data[frame_id][:, 0], data[frame_id][:, 2], cmap='Greens')

ax.set_xlabel('y')
ax.set_ylabel('x')
ax.set_zlabel('z')

fig = plt.figure()
plt.scatter(data[frame_id][:, 1], data[frame_id][:, 0])
plt.xlabel("y")
plt.ylabel("z")
```
